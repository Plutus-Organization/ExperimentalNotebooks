```python
from stock import get_history, get_intraday
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 7 day history for AAPL: get_history(ticker, num_days)
# Format is [datetimes], [prices]
get_history('AAPL', 7)
```

```python
plt.plot(timestamps, prices)
```
