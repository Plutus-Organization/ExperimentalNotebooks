```python
from stock import get_history, get_intraday, get_price_change
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 7 day history for AAPL: get_history(ticker, num_days)
# Format is [datetimes], [prices]
get_history('AAPL', 7)
```

```python
timestamps, prices = get_history('AAPL', 30)
```

```python
plt.title('AAPL 30 Day History')
plt.plot(timestamps, prices)
```

```python
# Intraday format is get_intraday(ticker, interval) 
timestamps, prices = get_intraday('AAPL', 5)
```

```python
plt.plot(timestamps, prices)
plt.title('AAPL 2017-07-14 Performance')
```

```python
# Day to day change
get_price_change('AAPL')
```
